```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

## Problem 1 Part C: Comment on two controller design

Assuming c=1

```python
x= np.linspace(0,10,100)

#CLF
u1= 2*x #abs value

#Sontag's Formula
u2= x+x*(np.sqrt(1+x**4)) #abs value
```

```python
plt.plot(x,u1, label= 'Lyapunov based controller')
plt.plot(x,u2, label = 'Sontags formula')
plt.legend(loc='best')
plt.xlabel('|x|')
plt.ylabel('|u|')
plt.savefig('HW-3 Problem 1')
```

# Part D: Closed-loop behaviour

c=1

Initial conditions:
1. x(0) = 1
2. x(0) = 2
3. x(0) = 10



```python
t=np.linspace(0,4)

# CLP

def f1(x, t):
    dxdt= -x**3
    return dxdt

x1= odeint(f1, 1, t)
x2= odeint(f1, 2, t)
x3= odeint(f1, 10, t)

plt.plot(t,x1, label= 'x(0)=1')
plt.plot(t,x2, label= 'x(0)=2')
plt.plot(t,x3, label= 'x(0)=10')
plt.legend(loc='best')
plt.title('Closed-loop behaviour for CLP')
plt.xlabel('time (s)')
plt.ylabel('x')
plt.xlim(0,4)
plt.ylim(0,10)
plt.savefig('Part D- CLP')
plt.show()




```

```python
#Sontag's formula

def f2(x,t):
    dxdt= -(x**3)*np.sqrt(1+x**4)
    return dxdt

x1= odeint(f2, 1, t)
x2= odeint(f2, 2, t)
x3= odeint(f2, 10, t)

plt.plot(t,x1, label= 'x(0)=1')
plt.plot(t,x2, label= 'x(0)=2')
plt.plot(t,x3, label= 'x(0)=10')
plt.legend(loc='best')
plt.title('Closed-loop design for controller using Sontags formula')
plt.xlabel('time (s)')
plt.ylabel('x')
plt.xlim(0,4)
plt.ylim(0,10)

plt.savefig('Part D- Sontag')
plt.show()
```
