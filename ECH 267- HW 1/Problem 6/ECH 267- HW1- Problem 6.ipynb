```python
import numpy as np
import matplotlib.pyplot as plt

def f(Y, t):
    y1, y2 = Y
    return [a11*y1+a12*y2, a21*y1+a22*y2]

y1 = np.linspace(-2.0, 2.0, 5)
y2 = np.linspace(-2.0, 2.0, 5)

Y1, Y2 = np.meshgrid(y1, y2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape


```

```python
#Problem 6(i)

#original Coordinates

a11=0
a12=1
a21=-2
a22=-3

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
     

Q = plt.streamplot(Y1, Y2, u, v, color='r')

plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.show()
plt.savefig('6(i) x')
```
