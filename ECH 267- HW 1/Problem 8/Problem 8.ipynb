```python
import numpy as np
import matplotlib.pyplot as plt
import math

```

```python
#Problem 8(i)

def f(X, t):
    x1, x2 = X
    return [x2, x1-2*math.atan(x1+x2) ]

x1 = np.linspace(-2.0, 2.0, 5)
x2 = np.linspace(-2.0, 2.0, 5)

X1, X2 = np.meshgrid(x1, x2)

t = 0

u, v = np.zeros(X1.shape), np.zeros(X2.shape)

NI, NJ = X1.shape

for i in range(NI):
    for j in range(NJ):
        x = X1[i, j]
        y = X2[i, j]
        xprime = f([x, y], t)
        u[i,j] = xprime[0]
        v[i,j] = xprime[1]
     

Q = plt.streamplot(X1, X2, u, v, color='r')

plt.title('Problem 8 (i)',fontsize=16)
plt.xlabel('$x_1$',fontsize=12)
plt.ylabel('$x_2$',fontsize=12)
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.show()
plt.savefig('8(i)')
```

```python
#Problem 8(ii)

def f(X, t):
    x1, x2 = X
    return [x2, -x1+x2*(1-3*x1**2-2*x2**2) ]

x1 = np.linspace(-2.0, 2.0, 5)
x2 = np.linspace(-2.0, 2.0, 5)

X1, X2 = np.meshgrid(x1, x2)

t = 0

u, v = np.zeros(X1.shape), np.zeros(X2.shape)

NI, NJ = X1.shape

for i in range(NI):
    for j in range(NJ):
        x = X1[i, j]
        y = X2[i, j]
        xprime = f([x, y], t)
        u[i,j] = xprime[0]
        v[i,j] = xprime[1]
     

Q = plt.streamplot(X1, X2, u, v, color='r')

plt.title('Problem 8 (ii)',fontsize=16)
plt.xlabel('$x_1$',fontsize=12)
plt.ylabel('$x_2$',fontsize=12)
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.show()
plt.savefig('8(ii)')
```
