```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Problem 10

def f(X, t):
    x1, x2 = X
    return [x2, -x1-x2 ]

x1 = np.linspace(-2.0, 2.0, 5)
x2 = np.linspace(-2.0, 2.0, 5)

X1, X2 = np.meshgrid(x1, x2)

t = 0

u, v = np.zeros(X1.shape), np.zeros(X2.shape)

NI, NJ = X1.shape

for i in range(NI):
    for j in range(NJ):
        x = X1[i, j]
        y = X2[i, j]
        xprime = f([x, y], t)
        u[i,j] = xprime[0]
        v[i,j] = xprime[1]
     

Q = plt.streamplot(X1, X2, u, v, color='r')

plt.xlabel('$x_1$',fontsize=12)
plt.ylabel('$x_2$',fontsize=12)
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.axis('square')
plt.savefig('10')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
x1 = np.linspace(-2.0, 2.0, 1000)
y = x1

[X, Y] = np.meshgrid(x1,y)
  
fig, ax = plt.subplots(1, 1)
  
Z= 0.75*X**2-(1/12)*X**4+0.5*X*Y+0.5*Y**2
  

CS = ax.contour(X, Y, Z)
plt.clabel(CS, colors = 'k', fmt = '%2.1f', fontsize=12)
  
ax.set_title('Contour for given lyapunov')
ax.set_xlabel('$x_{1}$')
ax.set_ylabel('$x_{2}$')
plt.savefig('Contour for lyapunov function') 
plt.show()
```
