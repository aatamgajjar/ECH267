```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,100)

x1= 1/np.sqrt(1+2*t)
x2= -x1
v=x1**2/2

plt.plot(t, x1, color = 'b', label = 'X(t)')
plt.plot(t, x2, color = 'b')
plt.plot(t, v, color = 'r', label = 'V(x(t))')
plt.legend(loc='best');
plt.xlabel('Time (s)')
plt.ylabel('V(x(t)) and X(t)')
plt.title('Comparison for V(x) and X(t) vs t')
plt.savefig('Problem 6')
plt.show()
```
